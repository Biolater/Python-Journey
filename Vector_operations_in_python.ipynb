```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale=1)
plt.quiver(0, 0, 3, -2, scale_units='xy', angles='xy', scale=1)
plt.quiver(0, 0, 5, 1, scale_units='xy', angles='xy', scale=1)
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
plt.quiver(0, 0, 5, 6, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, 3, -2, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
vector_1 = np.array([0, 0, 5, 6])
vector_2 = np.array([0, 0, 3, -2])
vector_sub = vector_1 - vector_2
print(vector_sub)
plt.quiver(0, 0, vector_sub[2], vector_sub[3], scale_units='xy', angles='xy', scale=1, color='g')
plt.show()
```

```python
vector_1 = np.array([0, 0, 2, 3])
vector_2 = 2 * vector_1
plt.quiver(0, 0, 2, 3, scale=1, scale_units='xy', angles='xy', color='b')
plt.quiver(0, 0, 4, 6, scale=1, scale_units='xy', angles='xy', color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
print(vector_2)
```

```python
vector_1 = np.array([0, 0, 2, 3])
vector_2 = -0.5 * vector_1
plt.quiver(0, 0, 2, 3, scale=1, scale_units='xy', angles='xy', color='b')
plt.quiver(0, 0, vector_2[2], vector_2[3], scale=1, scale_units='xy', angles='xy', color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
print(vector_2)
```
